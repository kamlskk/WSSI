##Przodek (jan, milena):
```
przodek(X, Y) :- ojciec(X, Y).
przodek(X, Y) :- ojciec(X, Z), przodek(Z, Y).
```
###Opis:

- Pierwsza część reguły mówi: `jan` jest przodkiem `milena`, jeśli `jan` jest jej ojcem. To jednak nie zachodzi bezpośrednio, ponieważ ojcem `milena` jest `grzegorz`.
-Druga część reguły działa rekursywnie: sprawdza, czy `jan` jest ojcem kogoś (w tym przypadku `elzbieta`), kto jest przodkiem `milena`.
-`jan` → ojciec `elzbieta`, a `elzbieta` → matka `milena`.
- W ten sposób `jan` zostaje uznany za przodka `milena`.

## Brat (grzegorz, tomasz).
```
    brat(X, Y) :- mezczyzna(X), ojciec(Z, X), ojciec(Z, Y), X \= Y.
    brat(X, Y) :- mezczyzna(X), matka(Z, X), matka(Z, Y), X \= Y.
```
###Opis:
- Reguła najpierw sprawdza, czy `grzegorz` i `tomasz` mają wspólnego ojca. Obaj mają ojca `stanislaw`.
- Następnie sprawdza, czy mają wspólną matkę. Obaj mają matkę `maria`.
- `grzegorz` jest mężczyzną (`mezczyzna(grzegorz)` zwraca `true`), a `grzegorz \= tomasz`.
- Wszystkie warunki są spełnione, więc `brat(grzegorz, tomasz)` zwraca `true`.

## Siostra (elzbieta, teresa).
```
    siostra(X, Y) :- kobieta(X), ojciec(Z, X), ojciec(Z, Y), X \= Y.
    siostra(X, Y) :- kobieta(X), matka(Z, X), matka(Z, Y), X \= Y.
```
###Opis:
- Reguła najpierw sprawdza, czy `elzbieta` i `teresa` mają wspólnego ojca. Obie mają ojca `jan`.
- Następnie sprawdza, czy mają wspólną matkę. Obie mają matkę `helena`.
- `elzbieta` jest kobietą (kobieta(elzbieta) zwraca true), a elzbieta \= teresa.
- Wszystkie warunki są spełnione, więc `siostra(elzbieta, teresa)` zwraca `true`.

## Kuzyn (szymon, jakub).
```
    kuzyn(X, Y) :- ojciec(A, X), ojciec(B, Y), brat(A, B), X \= Y.
```
###Opis:
- Ojcem `szymon` jest `grzegorz`, a ojcem `jakub` jest `tomasz`.
- Reguła sprawdza, czy `grzegorz` i `tomasz` są braćmi. Wynik dla `brat(grzegorz, tomasz)` zwrócił `true`.
- `szymon` i `jakub` to różne osoby ` (szymon \= jakub)`
- Wszystkie warunki są spełnione, więc `kuzyn(szymon, jakub)` zwraca `true`.

##**Sprawdzenie połączenia**
```
polaczenie(X, Y) :- polaczenie(X, Y, _).
```
###**Opis:**

- Sprawdza, czy istnieje połączenie między dwoma lotniskami X i Y.
- Używa istniejących faktów polaczenie(X, Y, C), gdzie C jest dowolny.
- Wynik to true, jeśli takie połączenie istnieje, inaczej false.

##**Określenie kosztu podróży**

```
koszt_podrozy(X, Y, C) :- polaczenie(X, Y, C).
```
###**Opis:**

- Znajduje koszt bezpośredniego połączenia między `X` i `Y`.
- Dopasowuje zmienną `C` do kosztu przechowywanego w fakcie `polaczenie(X, Y, C)`.
- Jeśli połączenie istnieje, zwraca wartość kosztu, inaczej zapytanie kończy się niepowodzeniem.

## **Wyznaczenie trasy (rekurencyjnie)**
```
trasa(X, Y, [X, Y], C) :- polaczenie(X, Y, C).
trasa(X, Y, [X | ResztaTrasy], C) :-
    polaczenie(X, Z, C1),
    Z \= Y,
    trasa(Z, Y, ResztaTrasy, C2),
    \+ member(X, ResztaTrasy),
    C is C1 + C2.
```
###**Opis:**

- Pierwsza reguła: Wyznacza trasę dla bezpośredniego połączenia między `X` i `Y`, zwracając `[X, Y]` jako trasę i koszt `C`.
- Druga reguła: Znajduje trasę wieloetapową przez pośrednie lotniska.
  - Dopasowuje lotnisko `Z` jako przystanek pośredni między `X` i `Y`.
  - Wywołuje rekurencyjnie siebie dla trasy `Z → Y`.
  - Sumuje koszty połączeń etapowych.
  - Zapobiega zapętleniu za pomocą `\+ member(X, ResztaTrasy)`.

## **Wyznaczenie najkrótszej trasy**

```
najkrotsza_droga(X, Y, Trasa, Koszt) :-
    findall([C, T], trasa(X, Y, T, C), Wyniki),
    sort(Wyniki, [[Koszt, Trasa] | _]).
```
###**Opis:**

- Zbiera wszystkie możliwe trasy z `X` do `Y` za pomocą `findall/3`.
- Każda trasa jest przechowywana w formacie `[Koszt, Trasa]`.
- Sortuje trasy według kosztu i zwraca pierwszą jako najkrótszą.
- Jeśli nie ma trasy, zapytanie kończy się niepowodzeniem.

##**Wyszukiwanie wszystkich połączeń**
```
wszystkie_polaczenia(X, Y, Trasa, Koszt) :-
    trasa(X, Y, Trasa, Koszt).
```
###**Opis:

- Znajduje wszystkie możliwe trasy z `X` do `Y`, w tym wieloetapowe.
- Wykorzystuje regułę `trasa/4` do generowania tras i przypisywania ich kosztów.
- Nie sortuje ani nie filtruje wyników.

##Wyznaczenie najtańszej trasy

```
najtansza_droga(X, Y, Trasa, Koszt) :-
    setof([C, T], trasa(X, Y, T, C), Wyniki),
    Wyniki = [[Koszt, Trasa] | _].
```
###**Opis:**

- Zbiera wszystkie trasy z `X` do `Y` w formacie `[Koszt, Trasa]`, korzystając z setof/3.
- Wyniki są automatycznie uporządkowane rosnąco według kosztu.
- Zwraca trasę o najmniejszym koszcie jako pierwszy element listy.
- Jeśli brak trasy, zapytanie kończy się niepowodzeniem.

##**Zbieranie połączeń z lotniska**

```
polaczenia_z_lotniska(X, Polaczenia) :-
    findall([Y, C], polaczenie(X, Y, C), Polaczenia).
```
###**Opis:**

- Zbiera wszystkie połączenia wychodzące z lotniska X w formie listy [Y, C], gdzie Y to cel, a C to koszt.
- Wykorzystuje findall/3 do wygenerowania listy wszystkich dopasowanych wyników.
- Wynik to lista par [cel, koszt].

##**Sprawdzanie bezpośredniego połączenia**

```
bezposrednie_polaczenie(X, Y) :-
    polaczenie(X, Y, _).
```
###**Opis:**

- Sprawdza, czy istnieje bezpośrednie połączenie między lotniskami `X` i `Y`.
