```python
miasto(warszawa).
miasto(krakow).
miasto(poznan).
miasto(wroclaw).
miasto(gdansk).
miasto(szczecin).
miasto(lublin).
miasto(bialystok).
miasto(katowice).
miasto(rzeszow).
miasto(opole).
miasto(torun).
miasto(legnica).
miasto(olsztyn).
miasto(koszalin).
miasto(gorzow).
miasto(zielona_gora).
miasto(suwalki).
miasto(mazury).
miasto(radom).

droga(warszawa, krakow, 5).
droga(krakow, warszawa, 5).
droga(krakow, poznan, 4).
droga(poznan, krakow, 4).
droga(warszawa, poznan, 3).
droga(poznan, warszawa, 3).
droga(wroclaw, warszawa, 5).
droga(warszawa, wroclaw, 5).
droga(gdansk, warszawa, 5).
droga(warszawa, gdansk, 5).
droga(szczecin, warszawa, 5).
droga(warszawa, szczecin, 5).
droga(lublin, warszawa, 2).
droga(warszawa, lublin, 2).
droga(bialystok, warszawa, 4).
droga(warszawa, bialystok, 4).
droga(krakow, wroclaw, 3).
droga(wroclaw, krakow, 3).
droga(krakow, gdansk, 5).
droga(gdansk, krakow, 5).
droga(krakow, szczecin, 6).
droga(szczecin, krakow, 6).
droga(poznan, wroclaw, 2).
droga(wroclaw, poznan, 2).
droga(poznan, gdansk, 5).
droga(gdansk, poznan, 5).
droga(poznan, szczecin, 4).
droga(szczecin, poznan, 4).
droga(krakow, opole, 3).
droga(opole, krakow, 3).
droga(krakow, torun, 3).
droga(torun, krakow, 3).
droga(warszawa, opole, 3).
droga(opole, warszawa, 3).
droga(warszawa, torun, 3).
droga(torun, warszawa, 3).
droga(warszawa, legnica, 4).
droga(legnica, warszawa, 4).
droga(warszawa, olsztyn, 4).
droga(olsztyn, warszawa, 4).
droga(krakow, legnica, 4).
droga(legnica, krakow, 4).
droga(wroclaw, opole, 1).
droga(opole, wroclaw, 1).
droga(gdansk, wroclaw, 6).
droga(wroclaw, gdansk, 6).
droga(gdansk, szczecin, 4).
droga(szczecin, gdansk, 4).
droga(szczecin, wroclaw, 5).
droga(wroclaw, szczecin, 5).
droga(lublin, bialystok, 2).
droga(bialystok, lublin, 2).
droga(lublin, katowice, 4).
droga(katowice, lublin, 4).
droga(rzeszow, bialystok, 4).
droga(bialystok, rzeszow, 4).
droga(opole, katowice, 3).
droga(katowice, opole, 3).
droga(torun, rzeszow, 4).
droga(rzeszow, torun, 4).
droga(legnica, katowice, 4).
droga(katowice, legnica, 4).
droga(olsztyn, warszawa, 4).
droga(warszawa, olsztyn, 4).
droga(gorzow, szczecin, 1).
droga(szczecin, gorzow, 1).
droga(gorzow, bialystok, 7).
droga(bialystok, gorzow, 7).
droga(zielona_gora, bialystok, 5).
droga(bialystok, zielona_gora, 5).
droga(koszalin, bialystok, 6).
droga(bialystok, koszalin, 6).

% Dodatkowe połączenia
droga(radom, warszawa, 1).
droga(warszawa, radom, 1).
droga(suwalki, bialystok, 2).
droga(bialystok, suwalki, 2).
droga(mazury, bialystok, 3).
droga(bialystok, mazury, 3).
droga(mazury, warszawa, 4).
droga(warszawa, mazury, 4).
droga(koszalin, szczecin, 2).
droga(szczecin, koszalin, 2).
droga(koszalin, wroclaw, 6).
droga(wroclaw, koszalin, 6).
droga(katowice, gorzow, 5).
droga(gorzow, katowice, 5).
droga(zielona_gora, katowice, 5).
droga(katowice, zielona_gora, 5).
droga(zielona_gora, rzeszow, 6).
droga(rzeszow, zielona_gora, 6).
droga(torun, koszalin, 5).
droga(koszalin, torun, 5).
droga(olsztyn, suwalki, 3).
droga(suwalki, olsztyn, 3).

% Reguła sprawdzająca, czy istnieje bezpośrednie połączenie między miastami X i Y
droga(X, Y) :- droga(X, Y, _).

% Reguła obliczająca czas przejazdu między dwoma miastami, uwzględniająca połączenia pośrednie
czas_przejazdu(X, Y, T) :-
    droga(X, Y, T). % Bezpośrednie połączenie
czas_przejazdu(X, Y, T) :-
    droga(X, Z, T1), % Połączenie do miasta pośredniego Z
    \+ member(Z, [X]), % Zapobiegamy powtarzaniu się miasta Z
    czas_przejazdu(Z, Y, T2), % Czas przejazdu od Z do Y
    (   T2 = infinity -> T = infinity % Jeśli brak dalszej trasy
    ;   T is T1 + T2 % Sumowanie czasu, jeśli istnieje dalsza trasa
    ).

% Reguła wyznaczająca najkrótszą trasę między dwoma miastami
najkrotsza_trasa(X, Y, NajkrotszaTrasa, NajkrotszyCzas) :-
    (   findall((Trasa, Czas), znajdz_trase(X, Y, Trasa, Czas), Trasy)
    ->  sort(2, @=<, Trasy, [(NajkrotszaTrasa, NajkrotszyCzas) | _])
    ;   NajkrotszaTrasa = [], NajkrotszyCzas = infinity % Brak dostępnych tras
    ).

% Pomocnicza reguła do znajdowania wszystkich możliwych tras
znajdz_trase(X, Y, [X, Y], Czas) :-
    droga(X, Y, Czas).
znajdz_trase(X, Y, [X | Trasa], Czas) :-
    droga(X, Z, Czas1),
    Z \= Y,
    \+ member(Z, [X | Trasa]), % Unikamy cykli
    znajdz_trase(Z, Y, Trasa, Czas2),
    (   Czas2 = infinity -> Czas = infinity
    ;   Czas is Czas1 + Czas2
    ).

% Reguła znajdowania wszystkich miast, które można osiągnąć z miasta X
miasta_dostepne(X, Miasta) :-
    findall(Y, (droga(X, Y, _); droga(Y, X, _)), MiastaUnikalne),
    list_to_set(MiastaUnikalne, Miasta).  % Usuń duplikaty, aby mieć unikalne miasta

% Dodawanie nowego połączenia do bazy danych
dodaj_droga(X, Y, Czas) :-
    assertz(droga(X, Y, Czas)),
    assertz(droga(Y, X, Czas)).

% Usuwanie połączenia z bazy danych
usun_droga(X, Y) :-
    retract(droga(X, Y, _)),
    retract(droga(Y, X, _)).

% Reguła sprawdzająca, czy droga działa w obie strony
droga_obie_strony(X, Y) :-
    droga(X, Y, _),
    droga(Y, X, _).

% Reguła znajdująca miasto z największą liczbą połączeń
miasto_z_najwieksza_liczba_polaczen(Miasto) :-
    findall(X, droga(X, _, _), Polaczenia1),
    findall(Y, droga(_, Y, _), Polaczenia2),
    append(Polaczenia1, Polaczenia2, Polaczenia),
    list_to_set(Polaczenia, UnikalneMiasta),
    max_list(UnikalneMiasta, Miasto).
```

```python
% Fakty dotyczące książek
ksiazka('W pustyni i w puszczy', sienkiewicz, przygodowa).
ksiazka('Hobbit', tolkien, fantasy).
ksiazka('Pan Tadeusz', mickiewicz, poezja).
ksiazka('Duma i uprzedzenie', austen, romans).
ksiazka('Władca pierścieni', tolkien, fantasy).
ksiazka('Zbrodnia i kara', dostojewski, kryminał).
ksiazka('Mistrz i Małgorzata', bułhakow, fantasy).
ksiazka('Pani Bovary', flaubert, romans).
ksiazka('Harry Potter i Kamień Filozoficzny', rowling, fantasy).
ksiazka('1984', orwell, dystopia).
ksiazka('Zabić drozda', lee, dramat).
ksiazka('Cień wiatru', zafon, thriller).
ksiazka('Przeminęło z wiatrem', mitchell, romans).
ksiazka('Wielki Gatsby', fitzgerald, literatura klasyczna).
ksiazka('Lalka', prus, powieść społeczna).
ksiazka('Moby Dick', melville, przygodowa).
ksiazka('Frankenstein', shelley, horror).
ksiazka('Sherlock Holmes', doyle, kryminał).
ksiazka('Wojna i pokój', tolstoj, historyczna).
ksiazka('Odyseja', homer, epika).
ksiazka('Złodziejka książek', zusak, dramat).

% Fakty dotyczące ocen
ocena(uzytkownik1, 'W pustyni i w puszczy', 5).
ocena(uzytkownik1, 'Hobbit', 4).
ocena(uzytkownik1, 'Pan Tadeusz', 3).
ocena(uzytkownik1, 'Duma i uprzedzenie', 5).
ocena(uzytkownik1, 'Władca pierścieni', 5).
ocena(uzytkownik1, 'Zbrodnia i kara', 4).
ocena(uzytkownik1, 'Mistrz i Małgorzata', 5).
ocena(uzytkownik1, 'Pani Bovary', 3).
ocena(uzytkownik1, 'Harry Potter i Kamień Filozoficzny', 5).
ocena(uzytkownik1, '1984', 4).
ocena(uzytkownik1, 'Zabić drozda', 5).
ocena(uzytkownik1, 'Cień wiatru', 4).
ocena(uzytkownik1, 'Przeminęło z wiatrem', 3).
ocena(uzytkownik1, 'Wielki Gatsby', 5).
ocena(uzytkownik1, 'Lalka', 4).
ocena(uzytkownik1, 'Moby Dick', 4).
ocena(uzytkownik1, 'Frankenstein', 3).
ocena(uzytkownik1, 'Sherlock Holmes', 5).
ocena(uzytkownik1, 'Wojna i pokój', 4).
ocena(uzytkownik1, 'Odyseja', 5).
ocena(uzytkownik1, 'Złodziejka książek', 5).

ocena(uzytkownik2, 'W pustyni i w puszczy', 4).
ocena(uzytkownik2, 'Hobbit', 5).
ocena(uzytkownik2, 'Pan Tadeusz', 2).
ocena(uzytkownik2, 'Duma i uprzedzenie', 3).
ocena(uzytkownik2, 'Władca pierścieni', 5).
ocena(uzytkownik2, 'Zbrodnia i kara', 5).
ocena(uzytkownik2, 'Mistrz i Małgorzata', 3).
ocena(uzytkownik2, 'Pani Bovary', 4).
ocena(uzytkownik2, 'Harry Potter i Kamień Filozoficzny', 5).
ocena(uzytkownik2, '1984', 4).
ocena(uzytkownik2, 'Zabić drozda', 3).
ocena(uzytkownik2, 'Cień wiatru', 5).
ocena(uzytkownik2, 'Przeminęło z wiatrem', 4).
ocena(uzytkownik2, 'Wielki Gatsby', 5).
ocena(uzytkownik2, 'Lalka', 3).
ocena(uzytkownik2, 'Moby Dick', 4).
ocena(uzytkownik2, 'Frankenstein', 5).
ocena(uzytkownik2, 'Sherlock Holmes', 3).
ocena(uzytkownik2, 'Wojna i pokój', 5).
ocena(uzytkownik2, 'Odyseja', 4).
ocena(uzytkownik2, 'Złodziejka książek', 2).

ocena(uzytkownik3, 'W pustyni i w puszczy', 3).
ocena(uzytkownik3, 'Hobbit', 5).
ocena(uzytkownik3, 'Pan Tadeusz', 4).
ocena(uzytkownik3, 'Duma i uprzedzenie', 5).
ocena(uzytkownik3, 'Władca pierścieni', 4).
ocena(uzytkownik3, 'Zbrodnia i kara', 3).
ocena(uzytkownik3, 'Mistrz i Małgorzata', 4).
ocena(uzytkownik3, 'Pani Bovary', 5).
ocena(uzytkownik3, 'Harry Potter i Kamień Filozoficzny', 4).
ocena(uzytkownik3, '1984', 5).
ocena(uzytkownik3, 'Zabić drozda', 4).
ocena(uzytkownik3, 'Cień wiatru', 5).
ocena(uzytkownik3, 'Przeminęło z wiatrem', 4).
ocena(uzytkownik3, 'Wielki Gatsby', 3).
ocena(uzytkownik3, 'Lalka', 5).
ocena(uzytkownik3, 'Moby Dick', 3).
ocena(uzytkownik3, 'Frankenstein', 4).
ocena(uzytkownik3, 'Sherlock Holmes', 4).
ocena(uzytkownik3, 'Wojna i pokój', 3).
ocena(uzytkownik3, 'Odyseja', 5).
ocena(uzytkownik3, 'Złodziejka książek', 4).

ocena(uzytkownik4, 'W pustyni i w puszczy', 5).
ocena(uzytkownik4, 'Hobbit', 4).
ocena(uzytkownik4, 'Pan Tadeusz', 4).
ocena(uzytkownik4, 'Duma i uprzedzenie', 4).
ocena(uzytkownik4, 'Władca pierścieni', 4).
ocena(uzytkownik4, 'Zbrodnia i kara', 4).
ocena(uzytkownik4, 'Mistrz i Małgorzata', 4).
ocena(uzytkownik4, 'Pani Bovary', 4).
ocena(uzytkownik4, 'Harry Potter i Kamień Filozoficzny', 4).
ocena(uzytkownik4, '1984', 4).
ocena(uzytkownik4, 'Zabić drozda', 4).
ocena(uzytkownik4, 'Cień wiatru', 4).
ocena(uzytkownik4, 'Przeminęło z wiatrem', 4).
ocena(uzytkownik4, 'Wielki Gatsby', 4).
ocena(uzytkownik4, 'Lalka', 4).
ocena(uzytkownik4, 'Moby Dick', 4).
ocena(uzytkownik4, 'Frankenstein', 4).
ocena(uzytkownik4, 'Sherlock Holmes', 4).
ocena(uzytkownik4, 'Wojna i pokój', 4).
ocena(uzytkownik4, 'Odyseja', 4).
ocena(uzytkownik4, 'Złodziejka książek', 4).

ocena(uzytkownik5, 'W pustyni i w puszczy', 5).
ocena(uzytkownik5, 'Hobbit', 5).
ocena(uzytkownik5, 'Pan Tadeusz', 5).
ocena(uzytkownik5, 'Duma i uprzedzenie', 5).
ocena(uzytkownik5, 'Władca pierścieni', 5).
ocena(uzytkownik5, 'Zbrodnia i kara', 5).
ocena(uzytkownik5, 'Mistrz i Małgorzata', 5).
ocena(uzytkownik5, 'Pani Bovary', 5).
ocena(uzytkownik5, 'Harry Potter i Kamień Filozoficzny', 5).
ocena(uzytkownik5, '1984', 5).
ocena(uzytkownik5, 'Zabić drozda', 5).
ocena(uzytkownik5, 'Cień wiatru', 5).
ocena(uzytkownik5, 'Przeminęło z wiatrem', 5).
ocena(uzytkownik5, 'Wielki Gatsby', 5).
ocena(uzytkownik5, 'Lalka', 5).
ocena(uzytkownik5, 'Moby Dick', 5).
ocena(uzytkownik5, 'Frankenstein', 5).
ocena(uzytkownik5, 'Sherlock Holmes', 5).
ocena(uzytkownik5, 'Wojna i pokój', 5).
ocena(uzytkownik5, 'Odyseja', 5).
ocena(uzytkownik5, 'Złodziejka książek', 5).

% Fakty dotyczące preferencji
preferencje(uzytkownik1, fantasy).
preferencje(uzytkownik1, romans).
preferencje(uzytkownik1, kryminał).

preferencje(uzytkownik2, fantasy).
preferencje(uzytkownik2, kryminał).
preferencje(uzytkownik2, dramat).

preferencje(uzytkownik3, romans).
preferencje(uzytkownik3, literatura_klasyczna).
preferencje(uzytkownik3, powieść_społeczna).

preferencje(uzytkownik4, fantasy).
preferencje(uzytkownik4, kryminał).
preferencje(uzytkownik4, dramat).

preferencje(uzytkownik5, fantasy).
preferencje(uzytkownik5, romans).
preferencje(uzytkownik5, literatura_klasyczna).

% Reguła do dynamicznego zarządzania ocenami
dodaj_ocene(Uzytkownik, Ksiazka, Ocena) :-
    assertz(ocena(Uzytkownik, Ksiazka, Ocena)).

usun_ocene(Uzytkownik, Ksiazka) :-
    retract(ocena(Uzytkownik, Ksiazka, _)).

% Zaktualizowanie oceny książki
zaktualizuj_ocene(Uzytkownik, Ksiazka, NowaOcena) :-
    usun_ocene(Uzytkownik, Ksiazka),
    dodaj_ocene(Uzytkownik, Ksiazka, NowaOcena).

% Reguła polecane_ksiazki/2
polecane_ksiazki(Uzytkownik, Gatunek) :-
    % Zbieranie książek, które pasują do gatunku i mają ocenę >= 4
    findall(Ksiazka,
            (ksiazka(Ksiazka, _, Gatunek),
             ocena(Uzytkownik, Ksiazka, Ocena),
             Ocena >= 4),
            Ksiazek),
    % Wyświetlanie wyników
    write('Polecane książki dla użytkownika '),
    write(Uzytkownik),
    write(' w gatunku '),
    write(Gatunek),
    write(': '),
    write(Ksiazek), nl.

% Polecane książki na podstawie preferencji użytkownika
polecane_ksiazki_na_podstawie_preferencji(Uzytkownik) :-
    % Zbieranie książek, które pasują do preferencji i mają ocenę >= 4
    findall(Ksiazka,
            (preferencje(Uzytkownik, Gatunek),
             ksiazka(Ksiazka, _, Gatunek),
             ocena(Uzytkownik, Ksiazka, Ocena),
             Ocena >= 4),
            Ksiazek),
    % Wyświetlanie wyników
    write('Polecane książki dla użytkownika '),
    write(Uzytkownik),
    write(' na podstawie preferencji: '),
    nl,
    maplist(writeln, Ksiazek).  % Wyświetlanie książek w nowych liniach

% Średnia ocena książki
srednia_ocena(Ksiazka, SredniaOcena) :-
    findall(Ocena, ocena(_, Ksiazka, Ocena), Oceny),
    length(Oceny, L),
    L > 0,  % Upewnij się, że książka ma co najmniej jedną ocenę
    sum_list(Oceny, Suma),
    SredniaOcena is Suma / L.

% Polecane książki na podstawie podobnych ocen
polecane_na_podstawie_ocen(Uzytkownik) :-
    findall(Ksiazka,
            (ocena(Uzytkownik, Ksiazka, Ocena),
             Ocena >= 4,
             ksiazka(Ksiazka, _, _)),
            Ksiazek),
    write('Polecane książki dla użytkownika '),
    write(Uzytkownik),
    write(' na podstawie ocen: '),
    nl,
    maplist(writeln, Ksiazek).

% Książki według gatunku i ocen
ksiazki_po_ocenie_i_gatunku(Gatunek) :-
    findall(Ksiazka,
            (ksiazka(Ksiazka, _, Gatunek),
             ocena(_, Ksiazka, Ocena),
             Ocena >= 4),
            Ksiazek),
    write('Książki w gatunku '),
    write(Gatunek),
    write(' z oceną >= 4: '),
    nl,
    maplist(writeln, Ksiazek).
```
